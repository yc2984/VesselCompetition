```python
import string
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#using the manually filled excel sheet, cause there're 7 vessels missing.
s3 = 'part3.csv'
printable = set(string.printable)
filter(lambda x: x in printable,s3)
df_more_length_info = pd.read_csv(s3)

df_more_length_info.head(15)
```

```python
df_more_length_info.info()
```

```python
#convert the type of year to float or datetime
df_more_length_info['Year_num'] = [int(x[:4]) for x in df_more_length_info['Year']]
df_more_length_info['Year_num'][:10]
```

```python
import datetime
df_more_length_info['Datetime'] = df_more_length_info['Year_num'].apply(lambda x: pd.to_datetime(x,format='%Y'))
df_more_length_info['Datetime']
```

```python
df_more_length_info.groupby('Year')['Year_number']
```

```python
plt.scatter(df_more_length_info['Year_num'],df_more_length_info['L(m)'])
```

```python
df_more_length_info['L(m)'].plot.box(title='DIstribution of vessel length',figsize=(16,8))
#plt.figsi
```

```python
df_more_length_info['L(m)'].describe()
```

```python
df_more_length_info.index = df_more_length_info['Datetime']
```

```python
df_more_length_info.head()
```

```python
top_10_countries = list(df_more_length_info.groupby('country')['L(m)'].count().nlargest(n=10).index)
```

```python

df_more_length_info[df_more_length_info['country'].isin(top_10_countries)][['L(m)','country']].boxplot(by='country',figsize=(16,8))
```

```python
df_more_length_info[df_more_length_info['L(m)']>120]
```

```python
# Select only the vessels > 120m
top_10_countries_120m_df = df_more_length_info[df_more_length_info['L(m)']>120]
```

```python
top_10_countries_120m_df[['L(m)','country']].groupby('country').count().plot(figsize=(16,8),kind='bar')
```
